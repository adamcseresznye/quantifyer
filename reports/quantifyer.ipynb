Welcome to Quantifyer! This application is designed to streamline the data analysis process for targeted mass-based LC/GC-MS applications, enabling rapid calculation of key metrics like recovery, correction factor, and concentration values. Additionally, it also provides informative plots to enable fast decision making and reduce turn around times.

::: {.callout-note}
You can access the project's app through its [Streamlit website](https://quantifyer.streamlit.app/).

:::

# How to use the App

1. Upload your CP-Seeker output files [here](https://cpseeker-postprocess.streamlit.app/).
2. Optionally adjust the confidence level, which is set to 80% by default.
3. Download the filtered and merged dataset.
